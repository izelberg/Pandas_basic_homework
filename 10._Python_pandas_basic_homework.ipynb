Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
# РЕШЕНИЕ ЗАДАЧИ 1
#Выбираем из датафрейма рейтингов данные из двух столбцов, для которых рейтинг равен 5
df_result = df_ratings[df_ratings.rating == 5][['movieId', 'rating']] 
#Записываем в новый датафрейм результат подсчета записей, сгруппированных по мувиИд, т.е. сколько записей с рейтингом 5 попало на каждый мувиИД
df_result2 = df_result.groupby('movieId').count()
#Записываем в новый датафрейм результат сортировки значений по убыванию подсчитанного количества записей мувиИД с рейтингом 5, по убыванию
df_result3 = df_result2.sort_values('rating', ascending=[False])
#Записываем в новый датафрейм одну первую строку с максимальным количеством записей фильма с макс. рейтингом.
df_result4 = df_result3.head(1)
#Записываем в новый датафрейм результат объединения предыдущего датафрейма с датафреймом муви, откуда берем только название фильма
df_result5 = pd.merge(df_movies, df_result4, how = 'inner', on = 'movieId')[['movieId', 'title', 'rating']]
df_result5
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df_power = pd.read_csv('power.csv')
df_power.head()
```

```python
#Ищем, как правильно пишется в данных название стран
df_power[ df_power['country'].str.contains('lith', case = False) ]['country'].unique()
```

```python
df_power[ df_power['country'].str.contains('latv', case = False) ]['country'].unique()
```

```python
df_power[ df_power['country'].str.contains('esto', case = False) ]['country'].unique()
```

```python
# РЕШЕНИЕ ЗАДАЧИ 2
# Делаем последовательные фильтры по заданным условиям, с записью каждого результата в отдельный датафрейм
df_power2 = df_power[ (df_power.quantity >= 0) ] 
df_power3 = df_power2[ (df_power2.country == 'Lithuania') | (df_power2.country == 'Latvia') | 
                     (df_power2.country == 'Estonia') ] 
df_power4 = df_power3[ (df_power3.category == 4) | (df_power3.category == 12) | (df_power3.category == 21) ]
df_power5 = df_power4[ (df_power4.year >= 2005) & (df_power4.year <= 2010) ] 
# Делаем группировку с суммированием количества по трем уровням: страна, год, категория. Такие данные могут пригодиться.
df_power6 = df_power5.groupby(['country', 'year', 'category']).sum()
# Делаем группировку с суммированием количества по странам по всем условиям.
df_power7 = df_power5.groupby('country').sum()[['quantity']]
print(df_power6)
print(df_power7)
```
